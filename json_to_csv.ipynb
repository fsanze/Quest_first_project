```python
import pandas as pd
import json

pd.set_option('display.max_colwidth', None)
```

```python
#cargamos el JSON
with open("data.json", "r") as file:
    data = json.load(file)
```

```python
len(data)
```

```python
data[0]
```

```python
if "Time Series (Daily)" in data[0]:
    time_series = data[0]["Time Series (Daily)"]
time_series
```

```python
ticker_list = []
for i in range(26):
    ticker_list.append(data[i]["Meta Data"]["2. Symbol"])
ticker_list
```

```python
#Creamos una lista vacia para ir guardando la información ya con la estructura que queremos
all_data = []

#Bucle para ir pasando por los datos de cada valor
for element in data:
    ticker = element["Meta Data"]["2. Symbol"]
    
    for key, value in element["Time Series (Daily)"].items():
        #creamos un diccionario con las columnas que vamos a querer en el df y se lo añadimos a la lista 
        all_data.append({
            "Date": key,
            "Symbol": ticker,
            "Open": float(element["Time Series (Daily)"][key]["1. open"]),
            "High": float(element["Time Series (Daily)"][key]["2. high"]),
            "Low": float(element["Time Series (Daily)"][key]["3. low"]),
            "Close": float(element["Time Series (Daily)"][key]["4. close"]),
            "Volume": int(element["Time Series (Daily)"][key]["5. volume"]),
            })

#Formamos el df y lo omprobamos
df = pd.DataFrame(all_data)
df
```

```python
#Guardamos los datos en CSVs
df.to_csv("historical_data_backup.csv")
df.to_csv("historical_data.csv", index=False)
```
