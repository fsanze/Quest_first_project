```python
top_sp500 = {
    "Apple Inc.": "AAPL",
    "Microsoft Corporation": "MSFT",
    "Alphabet Inc. (Class A)": "GOOGL",
    "Alphabet Inc. (Class C)": "GOOG",
    "Amazon.com, Inc.": "AMZN",
    "NVIDIA Corporation": "NVDA",
    "Berkshire Hathaway Inc.": "BRK.B",
    "Meta Platforms, Inc.": "META",
    "Tesla, Inc.": "TSLA",
    "Eli Lilly and Company": "LLY",
    "Visa Inc.": "V",
    "UnitedHealth Group Incorporated": "UNH",
    "Johnson & Johnson": "JNJ",
    "Exxon Mobil Corporation": "XOM",
    "JPMorgan Chase & Co.": "JPM",
    "Walmart Inc.": "WMT",
    "Mastercard Incorporated": "MA",
    "Procter & Gamble Company": "PG",
    "Broadcom Inc.": "AVGO",
    "Home Depot, Inc.": "HD",
    "Oracle Corporation": "ORCL",
    "Chevron Corporation": "CVX",
    "Merck & Co., Inc.": "MRK",
    "AbbVie Inc.": "ABBV",
    "The Coca-Cola Company": "KO",
    "PepsiCo, Inc.": "PEP",
    "Costco Wholesale Corporation": "COST",
    "Adobe Inc.": "ADBE",
    "Bank of America Corporation": "BAC",
    "Cisco Systems, Inc.": "CSCO",
    "Pfizer Inc.": "PFE",
    "Thermo Fisher Scientific Inc.": "TMO",
    "McDonald's Corporation": "MCD",
    "Accenture plc": "ACN",
    "Salesforce, Inc.": "CRM",
    "Comcast Corporation": "CMCSA",
    "Danaher Corporation": "DHR",
    "Linde plc": "LIN",
    "Abbott Laboratories": "ABT",
    "Netflix, Inc.": "NFLX",
    "Advanced Micro Devices, Inc.": "AMD",
    "Nike, Inc.": "NKE",
    "T-Mobile US, Inc.": "TMUS",
    "The Walt Disney Company": "DIS",
    "Wells Fargo & Company": "WFC",
    "Texas Instruments Incorporated": "TXN",
    "Philip Morris International Inc.": "PM",
    "United Parcel Service, Inc.": "UPS",
    "Morgan Stanley": "MS",
    "ConocoPhillips": "COP",
    "Amgen Inc.": "AMGN"
    }
```

```python
#sacamos una lista de tuplas con el ticker (importante para poder hacer luego el cruce en el df) y el link de google news 
all_links = []

for k,v in top_sp500.items():
    language = "en"
    country = "US"
    keywords = f'"{k}"'
    keywords = keywords.replace(" ", "%20")
    site = "reuters.com"
    start_date = "2023-01-01"
    end_date = "2023-12-31"
    link = f"https://news.google.com/rss/search?q={keywords}+after:{start_date}+before:{end_date}+site:{site}&hl={language}&gl={country}&ceid={country}:{language}"
    all_links.append((v,link))
all_links
```

```python
#Creamos una lista de diccionarios con el ticker de cada compañia y los titulares, fecha de publicación de los resultados de los scrapeos
all_google_news = []

for link in all_links:
    response = requests.get(link[1])
    html = response.text
    soup = BeautifulSoup(html, 'html.parser')
    news = soup.find_all("item")
    for element in news:
        all_google_news.append({
            "Date": element.find("pubdate").get_text()[5:16],
            "Symbol": link[0],
            "News": element.find("title").get_text()})
```

```python
all_google_news
```

```python
#Creamos el df
news_df = pd.DataFrame(data=all_google_news)
news_df.head()
```

```python
#Pasamos las fechas de string a formato fecha
news_df["Date"] = pd.to_datetime(news_df['Date'])
news_df.head()
```

```python
#Ordenamos por fecha de publicación de la noticia
news_df = news_df.sort_values(by=["Symbol", "Date"], ascending=(True, True)).reset_index(drop=True)
news_df
```

```python
#Quitamos noticias anteriores a 2023
news_df = news_df[news_df["Date"] >= "2023-01-01"]
news_df
```

```python
# Creamos el ID con la fecha y el ticker para poder hacer el merge sin problemas
news_df["ID"] = news_df["Date"].astype(str) + " " + news_df["Symbol"]
news_df
```

```python
#Guardamos el df en un CSV
news_df.to_csv("All_news_2023.csv", index=False)
```
