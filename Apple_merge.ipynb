```python
import pandas as pd

# ⚙️ Settings
pd.set_option('display.max_columns', None) # display all columns
```

```python
#Cargamos los CSV en sus df
df1 = pd.read_csv("Apple_2023.csv")
df2 = pd.read_csv("Apple_news_2023.csv")
```

```python
# Hacemos un merge outer para no perder las noticias de los fines de semana
df = df1.merge(df2, on="Date", how="outer")
df
```

```python
#Rellenamos los nans de la columna de News
df["News"] = df["News"].fillna("No leaked news")
df
```

```python
#Sacamos los 10 días con mayores subidas
df.nlargest(10, "Pct_Change")
```

```python
#Sacamos los 10 días con mayores caídas
df.nsmallest(10, "Pct_Change")
```
