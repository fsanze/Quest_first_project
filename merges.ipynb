```python
import pandas as pd

pd.set_option('display.max_colwidth', None)
```

```python
#Cargamos los CSV, hacemos el merge por el ID y comprobamos
df1 = pd.read_csv("Only_2023.csv")
df2 = pd.read_csv("All_news_2023.csv")
df = df1.merge(df2, on="ID", how="outer")
df
```

```python
#Rellenamos los nans de las columnas News, Date_x y Symbol_x (serán las principales)
df["News"] = df["News"].fillna("No news") #Rellenamos con "No news"
df["Date_x"] = df["Date_x"].fillna(df["Date_y"]) #Rellenamos con la fecha de la noticia (columna Date_y)
df["Symbol_x"] = df["Symbol_x"].fillna(df["Symbol_y"]) #Rellenamos con el ticker de la noticia (columna Symbol_y)
df = df.drop(columns=["Date_y",	"Symbol_y"]) #Quitamos ls columnas que ya no queremos
df
```

```python
#Renombramos las columnas de Dates y Symbols
df = df.fillna(0)
df.rename(columns={"Date_x": "Date", "Symbol_x": "Symbol"}, inplace=True)
```

```python
#Ordenamos por ticker y fecha
df = df.sort_values(by=['Symbol', 'Date']).reset_index(drop=True)
df
```

```python
#Mostramos las 10 mayores subidas
df.nlargest(10, "Pct_Change")
```

```python
#Mostramos las 10 mayores subidas
df.nsmallest(10, "Pct_Change")
```

```python
# Sacamos el Top 10 de día con la mayor varíación positiva por cada compañia:
df.loc[df.groupby('Symbol')['Pct_Change'].idxmax()].sort_values(by="Pct_Change", ascending=False).head(10)
```

```python
# Sacamos el Top 10 de día con la mayor varíación negativa por cada compañia:
df.loc[df.groupby('Symbol')['Pct_Change'].idxmin()].sort_values(by="Pct_Change", ascending=True).head(10)
```
